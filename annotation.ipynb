```python
def get_image(file):
    img = cv2.imread(file)
    if img.shape[0] < img.shape[1]:
        img = np.rot90(np.rot90(np.rot90(img)))
        cv2.imwrite("./rtmp.jpg", img)
        img = cv2.imread("./rtmp.jpg")
    return img
```

```python
def show(img, change_color = True):
    if change_color:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img)
```

```python
#--- performs Otsu threshold ---
def threshold(img, st):
    ret, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    cv2.imwrite(os.path.join(os.getcwd(), 'res_' + str(st) + '.jpg'), thresh) 
    return  thresh

```
